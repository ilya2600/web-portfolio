```python
# Модули
import random as r
import time as t

# ХП Игрока
playerHP = 100
playerName = input("Ваше имя:")
playerDMG = 25

# ХП Врага
enemyHP = 100
enemyDMG = 10

# Список Слов
words = ["python","javascript","html","css"]

enemyImg = '''
╬╬╬╬╬╬╬╬╬╬          ╬╬╬╬╬╬╬╬╬╬
╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬
╬||||||||||||||||||||||||||||╬
╬||||||||||||||||||||||||||||╬
╬||||*****||||||||||*****||||╬
╬||||*****||||||||||*****||||╬
╬||||||||||||||||||||||||||||╬
╬|||||||||**********|||||||||╬
╬|||||||||**|*|*|*|*|||||||||╬
╬|||||||||**********|||||||||╬
╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬╬
╬╬╬╬╬╬╬╬╬╬          ╬╬╬╬╬╬╬╬╬╬
╬||||||||╬          ╬||||||||╬
╬||||||||╬          ╬||||||||╬
╬||||||||╬          ╬||||||||╬
╬╬╬╬╬╬╬╬╬╬          ╬╬╬╬╬╬╬╬╬╬
'''

while enemyHP > 0 and playerHP > 0:
  wordIndex = r.randint(0,3)
  chosenWord = words[wordIndex]

  # Интерфейс
  print(enemyImg)
  print(playerName,"ХП:",playerHP)
  print("Враг ХП:",enemyHP)

  print("Слово:", chosenWord)
  startTime = t.time()

  # Ход Игрока
  playerInput = input("Пишите: ")
  endTime = t.time()
  passedTime = endTime - startTime

  playerHIT = (playerDMG - passedTime) * (playerInput == chosenWord)
  enemyHP = enemyHP - playerHIT

  # Ход Врага
  playerHP = playerHP - 45

  if enemyHP <= 0:
    print("Игрок Победил!")

  if playerHP <= 0:
    print("Враг Победил!")
```
